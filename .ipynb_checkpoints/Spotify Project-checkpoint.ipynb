```python
def extract_uri(query):
    query=input("Please enter a song: ")
    results=spotify.search(q=query, limit=1)
    uri=results["tracks"]["items"][0]["uri"]
    print(uri)
    print(spotify.audio_features(uri))
```

```python
def song_recommender(enter_song):
    while True:
        enter_song= input("Please enter a song you like: ")
        in100 = any(hot100.Title == enter_song)
        if in100 is True:
            print("Then you may as well like: ")
            print(hot100[["Title","Artist"]].sample())
            continue
        else:
            results=spotify.search(q=enter_song, limit=1)
            uri=results["tracks"]["items"][0]["uri"]
            features_song=spotify.audio_features(uri)
            print("Your song is not on the Billboard Hot100, please wait for your recommendation")
            features_song_df=pd.DataFrame(features_song)
            features_song_df_=features_song_df.iloc[:,:11]
            features_song_scaled=scaler.transform(features_song_df_)
            features_song_scaled_df=pd.DataFrame(features_song_scaled)
            user_clusters=kmeans.predict(features_song_scaled_df)
            final_cluster=user_clusters[0]
            print(big_df.loc[big_df["Cluster"]==final_cluster].sample())

            #search for the predict list and return a song of the same cluster
            #predict de la cancion

            break
```

```python
song_recommender(enter_song)
```
