```python
def get_playlist_tracks(playlist_URI):
    global tracks
    global long_names
    results = spotify.user_playlist_tracks(user="spotify",playlist_id=playlist_URI)
    tracks = results['items']
    while results['next']:
        results = spotify.next(results)
        tracks.extend(results['items'])
    long_names = []
    for item in tracks:
        long_names.append(item["track"]["name"])
    return tracks, long_names
```

```python
def playlist_to_uri_song(playlist):
    results = spotify.user_playlist_tracks(user="spotify", playlist_id=playlist)
    uris=[]
    features=[]
    titles=[]
    artists=[]
    tracks=results['items']

    
    while results['next']:
        results = spotify.next(results)
        tracks.extend(results['items'])
        
    for song in tracks:
        uris.append(song["track"]["uri"])
        titles.append(song["track"]["name"])
        artists_in_song=[]
        for artist in song["track"]["artists"]:
            artists_in_song.append(artist["name"])
        artists.append(artists_in_song)

    for uri in uris:
        features.append(spotify.audio_features(uri))

    a=pd.DataFrame({"title":titles,
                                "artist":artists,
                                "uri":uris,
                                  "features":features})

    c = list(spotify.audio_features("spotify:track:2MvvoeRt8NcOXWESkxWn3g")[0].keys())
    b = pd.DataFrame([list(a["features"].apply(lambda x: x[0].values()).loc[i]) for i in range(a.shape[0])], columns = c)
    b=b.drop(columns=['type', 'track_href', 'analysis_url'])
    d=pd.concat([a,b], axis = 1)
    d=d.drop(columns='features')

    return d
    
```

### Create a database of songs with their audio features

```python
def playlist_to_uri_song(playlist):
    playlist_data = spotify.user_playlist_tracks(user="spotify", playlist_id=playlist)
    uris=[]
    features=[]
    titles=[]
    artists=[]

    for song in playlist_data["items"]:
        uris.append(song["track"]["uri"])
        titles.append(song["track"]["name"])
        artists_in_song=[]
        for artist in song["track"]["artists"]:
            artists_in_song.append(artist["name"])
        artists.append(artists_in_song)
        
    for uri in uris:
        features.append(spotify.audio_features(uri))

    a=pd.DataFrame({"title":titles,
                              "artist":artists,
                              "uri":uris,
                              "features":features})
    
    c = list(spotify.audio_features("spotify:track:2MvvoeRt8NcOXWESkxWn3g")[0].keys())
    b = pd.DataFrame([list(a["features"].apply(lambda x: x[0].values()).loc[i]) for i in range(a.shape[0])], columns = c)
    b=b.drop(columns=['type', 'track_href', 'analysis_url'])
    d=pd.concat([a,b], axis = 1)
    d=d.drop(columns='features')
    
    return d
    
```
